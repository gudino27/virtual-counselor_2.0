# 10 — Load and Explore Benchmark Datasets
Load GSM8K (500-subset) and MATH (200-subset), inspect statistics, and confirm cache files exist.

```python
import sys, json, os
from pathlib import Path

REPO_ROOT = Path("../..")
sys.path.insert(0, str(REPO_ROOT / "src"))

from evaluation.benchmarks import BenchmarkLoader

loader = BenchmarkLoader(cache_dir=REPO_ROOT / "data" / "benchmarks")
```

## Load Datasets

```python
gsm8k = loader.load_gsm8k()
math = loader.load_math()
print(f"GSM8K records : {len(gsm8k)}")
print(f"MATH  records : {len(math)}")
```

## Dataset Statistics

```python
def stats(dataset, name):
    q_lens = [len(r["question"]) for r in dataset]
    a_lens = [len(r["answer"]) for r in dataset]
    print(f"\n{name}")
    print(f"  Mean question length : {sum(q_lens)/len(q_lens):.1f} chars")
    print(f"  Mean answer length   : {sum(a_lens)/len(a_lens):.1f} chars")
    print(f"  Min/Max question len : {min(q_lens)} / {max(q_lens)}")

stats(gsm8k, "GSM8K")
stats(math,  "MATH")
```

## Confirm Cache Files on Disk

```python
cache_dir = REPO_ROOT / "data" / "benchmarks"
for fname in ["gsm8k_subset.json", "math_subset.json"]:
    p = cache_dir / fname
    if p.exists():
        count = len(json.loads(p.read_text()))
        print(f"✓ {fname} exists — {count} records ({p.stat().st_size // 1024} KB)")
    else:
        print(f"✗ {fname} MISSING")
```
